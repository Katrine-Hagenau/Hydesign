```python
import matplotlib.pyplot as plt

# Data from user's table
curtailment = [17.115, 22.710, 45.796, 60.820, 66.189]   # GWh
npv_capex_ratio=[0.315, 0.388, 0.359, 0.357, 0.356]
labels = ['W3_S1', 'W3_S2', 'W3_S3', 'W3_S4', 'W3_S5']


plt.figure(figsize=(6, 4))
plt.scatter(curtailment, npv_capex_ratio)
for x, y, label in zip(curtailment, npv_capex_ratio, labels):
    plt.annotate(label, (x, y), textcoords="offset points", xytext=(5, 5), ha='left')

plt.xlabel('Curtailment (GWh)')
plt.ylabel('NPV/CAPEX')
plt.title('Curtailment vs NPV/CAPEX for HPP_W scenarios')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data from user's table
npv_capex = [0.315, 0.388, 0.359, 0.357, 0.356]
aep = [856.758, 906.006,991.806,1018.159,1025.856 ]    # GWh
curtailment =[17.115, 22.710, 45.796, 60.820, 66.189]
labels = ['W3_S1', 'W3_S2', 'W3_S3', 'W3_S4', 'W3_S5']


# Plot 1: AEP vs NPV/CAPEX
plt.figure(figsize=(6, 4))
plt.scatter(aep,npv_capex)
for x, y, label in zip(aep,npv_capex, labels):
    plt.annotate(label, (x, y), textcoords="offset points", xytext=(5, 5), ha='left')
plt.ylabel('NPV/CAPEX ')
plt.xlabel('AEP (GWh)')
plt.title('AEP vs NPV/CAPEX for HPP_W scenarios')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 2: AEP vs Total Curtailment
plt.figure(figsize=(6, 4))
plt.scatter(curtailment, aep)
for x, y, label in zip(curtailment, aep, labels):
    plt.annotate(label, (x, y), textcoords="offset points", xytext=(5, 5), ha='left')
plt.xlabel('Total Curtailment (GWh)')
plt.ylabel('AEP (GWh)')
plt.title('AEP vs Total Curtailment for HPP_W scenarios')
plt.grid(True)
plt.tight_layout()
plt.show()
```
